## Parte I — Lista de Adjacência

### 1.1 O que é?

A **lista de adjacência** é uma das formas mais usadas de representar grafos em programação. A ideia é simples: para cada vértice, mantemos uma **lista com todos os seus vizinhos diretos**.

Essa estrutura responde principalmente à pergunta:

> **"Quem são os vizinhos de um determinado vértice?"**

### 1.2 Como funciona na prática?

Em Python, usamos um **dicionário** onde:
- Cada **chave** é um vértice;
- Cada **valor** é uma lista com os vértices adjacentes.

**Exemplo com um grafo menor:**

```
Grafo:  X — Y — Z
            |
            W
```

Sua lista de adjacência seria:

```python
{
    'X': ['Y'],
    'Y': ['X', 'Z', 'W'],
    'Z': ['Y'],
    'W': ['Y']
}
```

Note que, como o grafo é **não direcionado**, toda aresta aparece **duas vezes**: uma para cada extremidade.

### 1.3 Lógica de construção

Para construir a lista de adjacência a partir de uma lista de arestas:

```
Para cada aresta (u, v):
    adicionar v à lista de u
    adicionar u à lista de v  ← necessário por ser não direcionado
```

## Parte II — Matriz de Adjacência

### 2.1 O que é?

A **matriz de adjacência** é uma tabela quadrada de tamanho **V × V**, onde V é o número de vértices. Cada célula `M[i][j]` indica se existe uma aresta entre o vértice `i` e o vértice `j`:

- **1** → existe aresta;
- **0** → não existe aresta.

Essa estrutura responde principalmente à pergunta:

> **"Dois vértices específicos estão diretamente conectados?"**

### 2.2 Como ler a matriz?

Para o grafo menor `X — Y — Z`:

```
       X   Y   Z
  X  [ 0,  1,  0 ]
  Y  [ 1,  0,  1 ]
  Z  [ 0,  1,  0 ]
```

- A linha de **Y** tem `1` nas colunas X e Z → Y se conecta a X e Z.
- A célula `M[X][Z] = 0` → X e Z não estão diretamente conectados.

### 2.3 Lógica de construção

```
1. Crie um mapa: vértice → índice numérico
2. Crie uma matriz n×n de zeros
3. Para cada aresta (u, v):
       M[índice[u]][índice[v]] = 1
       M[índice[v]][índice[u]] = 1  ← não direcionado
```

### 2.4 Propriedade importante — Simetria

Em grafos **não direcionados**, a matriz de adjacência é **sempre simétrica**:

```
M[i][j] == M[j][i]  para todo i, j
```

Isso acontece porque se existe aresta de u para v, também existe de v para u.

````python
# =============================================================
# PARTE II — Implemente a Matriz de Adjacência
# =============================================================
# Instrução: construa a matriz de adjacência usando apenas listas
# Python aninhadas (lista de listas). Não use numpy ainda.

def construir_matriz_adjacencia(vertices, arestas, direcionado=False):
    """
    Constrói a matriz de adjacência.

    Retorna:
        (indice, matriz) onde:
        - indice: dict mapeando vértice → posição na matriz
        - matriz: lista de listas n×n com 0s e 1s
    """
    n = len(vertices)
    indice = {v: i for i, v in enumerate(vertices)}

    # TODO: crie a matriz de zeros e preencha com as arestas
    matriz = [[0] * n for _ in range(n)]

    for u, v in arestas:
        matriz[indice[u]][indice[v]] = 1
        if not direcionado:
            matriz[indice[v]][indice[u]] = 1

    return indice, matriz


def imprimir_matriz_adj(vertices, indice, matriz):
    cabecalho = "     " + "  ".join(f"{v:>2}" for v in vertices)
    print(cabecalho)
    print("     " + "---" * len(vertices))
    
    for v in vertices:
        i = indice[v]
        linha = "  ".join(f"{x:>2}" for x in matriz[i])
        print(f"  {v} | {linha}")

# Teste
indice_adj, mat_adj = construir_matriz_adjacencia(vertices, arestas)

print("Matriz de Adjacência:")
imprimir_matriz_adj(vertices, indice_adj, mat_adj)
````

## Parte III — Matriz de Incidência

### 3.1 O que é?

A **matriz de incidência** é diferente das anteriores: aqui relacionamos **arestas com vértices**, e não vértices com vértices.

Sua dimensão é **E × V**, onde:
- **E** = número de arestas (linhas);
- **V** = número de vértices (colunas).

Essa estrutura responde principalmente à pergunta:

> **"Quais vértices participam de cada aresta?"**

### 3.2 Como ler a matriz?

Para o grafo `X — Y — Z` com arestas `e1=(X,Y)` e `e2=(Y,Z)`:

```
        X   Y   Z
  e1  [ 1,  1,  0 ]   ← e1 toca X e Y
  e2  [ 0,  1,  1 ]   ← e2 toca Y e Z
```

Cada linha tem **exatamente dois valores 1** em um grafo simples não direcionado.

### 3.3 Lógica de construção

```
1. Rotule cada aresta: e1, e2, e3, ...
2. Crie mapa: aresta → índice de linha
3. Crie mapa: vértice → índice de coluna
4. Crie matriz de zeros (E linhas × V colunas)
5. Para cada aresta eᵢ = (u, v):
       M[i][índice[u]] = 1
       M[i][índice[v]] = 1
```

### 3.4 Diferença para grafos direcionados

Em **dígrafos**, uma convenção comum é:
- **-1** no vértice de **origem** (saída);
- **+1** no vértice de **destino** (entrada).

Isso permite distinguir o sentido da aresta.

4.Em algoritmos de fluxo de redes ou quando as arestas possuem propriedades próprias (como pesos ou identificações únicas) que precisam ser mapeadas individualmente.

````python
# =============================================================
# PARTE III — Implemente a Matriz de Incidência
# =============================================================
# Instrução: rotule as arestas como 'e1', 'e2', ... e construa
# a matriz de incidência (E linhas × V colunas).

def construir_matriz_incidencia(vertices, arestas, direcionado=False):
    """
    Constrói a matriz de incidência.

    Retorna:
        (rotulos_arestas, indice_v, matriz) onde:
        - rotulos_arestas : lista de nomes das arestas ['e1','e2',...]
        - indice_v        : dict vértice → índice de coluna
        - matriz          : lista de listas E×V
    """
    rotulos = [f"e{i+1}" for i in range(len(arestas))]
    indice_v = {v: j for j, v in enumerate(vertices)}

    n_e = len(arestas)
    n_v = len(vertices)

    # TODO: crie a matriz de zeros e preencha com as incidências
    matriz = [[0] * n_v for _ in range(n_e)]

    for i, (u, v) in enumerate(arestas):
        idx_u = indice_v[u]
        idx_v = indice_v[v]
        
        if direcionado:
            matriz[i][idx_u] = -1
            matriz[i][idx_v] = 1
        else:
            matriz[i][idx_u] = 1
            matriz[i][idx_v] = 1

    return rotulos, indice_v, matriz


def imprimir_matriz_inc(vertices, rotulos, indice_v, matriz):

    cabecalho = "      " + "  ".join(f"{v:>2}" for v in vertices)
    print(cabecalho)
    print("      " + "----" * len(vertices))
    for i, rot in enumerate(rotulos):
        linha = "  ".join(f"{x:>2}" for x in matriz[i])
        print(f"  {rot:>3} | {linha}")


# Teste
rotulos, indice_v, mat_inc = construir_matriz_incidencia(vertices, arestas)

print("Matriz de Incidência:")
imprimir_matriz_inc(vertices, rotulos, indice_v, mat_inc)
````
